```python
df_silver = spark.table('crypto.crypto_schema.bronze_crypto')

df_silver.display()
```

```python
from pyspark.sql.functions import col, sum as spark_sum
df_silver = df_silver.drop(col('roi'),col('image'))

df_silver.display()
```

```python
from pyspark.sql.functions import col, sum as spark_sum

null_counts = df_silver.select([spark_sum(col(c).isNull().cast('int')).alias(c) for c in df_silver.columns])
display(null_counts)
```

```python
from pyspark.sql.functions import expr,col

df_silver = df_silver\
            .withColumn('last_updated',col('last_updated').cast('timestamp'))\
            .withColumn('atl_date',col('atl_date').cast('timestamp'))\
            .withColumn('ath_date',col('ath_date').cast('timestamp'))
df_silver.display()
```

```python
df_silver.printSchema()
```

```python
df_silver = df_silver.fillna(0,subset=['max_supply'])

df_silver.display()
```

```python
df_silver.write.format('delta').mode('overwrite').saveAsTable('crypto.crypto_schema.silver_crypto')
```
