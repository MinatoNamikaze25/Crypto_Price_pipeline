```python
import requests,pandas as pd 

response = requests.get('https://api.coingecko.com/api/v3/ping')
print(response.status_code)
print(response.json())
```

```python
from pyspark.sql.functions import *
data = requests.get('https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=100&page=1')

print(data.status_code)
crypto_data = data.json()

pandas_df = pd.DataFrame(crypto_data)

df_raw = spark.createDataFrame(pandas_df)
df_raw.printSchema()
df_raw.count()
```

```python
df_raw.filter(col('roi').isNotNull()).count()
```

```python
%sql
create catalog if not exists crypto;
create schema if not exists crypto.crypto_schema;
```

```python
%sql
use crypto.crypto_schema;
```

```python
df_raw.write.format('delta').mode('overwrite').saveAsTable('crypto.crypto_schema.bronze_crypto')
```
