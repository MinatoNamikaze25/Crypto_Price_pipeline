```python
from pyspark.sql.functions import col,round

df_gold_4 = spark.table('crypto.crypto_schema.silver_crypto')
df_gold_4 = df_gold_4.withColumn('price_change_percentage_24h',round(col('price_change_percentage_24h'),2))
df_gold_4 = df_gold_4.select(col('name'),col('current_price'),col('price_change_percentage_24h')).orderBy(col('price_change_percentage_24h').desc()).limit(10)

display(df_gold_4)

df_gold_4.write.format('delta').mode('overwrite')\
               .saveAsTable('crypto.crypto_schema.gold_top_gainers')
```

```python
df_gold_5 = spark.table('crypto.crypto_schema.silver_crypto')

df_gold_5 = df_gold_5.withColumn('price_change_percentage_24h',round(col('price_change_percentage_24h'),2))
df_gold_5 = df_gold_5.select(col('name'),col('current_price'),col('price_change_percentage_24h')).orderBy(col('price_change_percentage_24h').asc()).limit(10)

display(df_gold_5)

df_gold_5.write.format('delta').mode('overwrite')\
               .saveAsTable('crypto.crypto_schema.gold_top_losers')
```

```python
df_gold_6 = spark.table('crypto.crypto_schema.silver_crypto')

df_gold_6 = df_gold_6.withColumn('volatility',round(col('high_24h')-col('low_24h'),2)).select(col('name'),col('high_24h'),col('low_24h'),col('volatility'))\
                     .orderBy(col('volatility').desc()).limit(10)

df_gold_6.display()

df_gold_6.write.format('delta').mode('overwrite')\
         .saveAsTable('crypto.crypto_schema.gold_volatility')
```
