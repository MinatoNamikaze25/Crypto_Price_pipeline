```python
from pyspark.sql.functions import col,round

df_gold_7 = spark.table('crypto.crypto_schema.silver_crypto')

df_gold_7=df_gold_7.filter(col("ath_change_percentage")<-5).select(col('name'),col('current_price'),col('ath'),round(col('ath_change_percentage'),2).alias("ath_change_percentage")).orderBy(col('ath_change_percentage').desc()).limit(10)

display(df_gold_7)

df_gold_7.write.format('delta').mode('overwrite')\
         .saveAsTable('crypto.crypto_schema.gold_closest_to_ath')
```

```python
df_gold_8 = spark.table('crypto.crypto_schema.silver_crypto')

df_gold_8=df_gold_8.select(col('name'),col('current_price'),col('ath'),round(col('ath_change_percentage'),2).alias("ath_change_percentage")).orderBy(col('ath_change_percentage').asc()).limit(10)

display(df_gold_8)

df_gold_8.write.format('delta').mode('overwrite')\
         .saveAsTable('crypto.crypto_schema.gold_furthest_to_ath')
```
