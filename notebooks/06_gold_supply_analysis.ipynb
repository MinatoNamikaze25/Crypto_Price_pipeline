```python
from pyspark.sql.functions import col,count,when

df_gold_11 = spark.table('crypto.crypto_schema.silver_crypto')

df_gold_11 = df_gold_11.withColumn('supply_type',(when(col('max_supply')>0,'limited').otherwise('unlimited')))

df_gold_11 = df_gold_11.groupBy("supply_type").agg(count("*").alias('coin_count'))

df_gold_11.display()

df_gold_11.write.format('delta').mode('overwrite').saveAsTable('crypto.crypto_schema.gold_supply_type')
```

```python
from pyspark.sql.functions import round,col
df_gold_12 = spark.table('crypto.crypto_schema.silver_crypto')

df_gold_12 = df_gold_12.withColumn('circulating_pct',round((col('circulating_supply')/col('max_supply'))*100,2))

df_gold_12 = df_gold_12.filter(col('max_supply')>0).select(col('name'),col('circulating_supply'),col('total_supply'),col('max_supply'),col('circulating_pct'))\
                       .orderBy(col('circulating_pct').desc()).limit(10)
df_gold_12.display()

df_gold_12.write.format('delta').mode('overwrite').saveAsTable('crypto.crypto_schema.gold_circulating_supply_pct')
```
