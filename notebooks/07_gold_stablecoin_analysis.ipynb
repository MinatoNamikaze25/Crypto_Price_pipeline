```python
from pyspark.sql.functions import when,col
df_gold_13 = spark.table('crypto.crypto_schema.silver_crypto')

df_gold_13 = df_gold_13.withColumn('is_stable',when((col('current_price')>=0.95)&(col('current_price')<=1.05),"stable").otherwise('Unstable'))

df_gold_13 = df_gold_13.filter(col('is_stable')=='stable').select(col("name"),col('current_price'),col('market_cap'),col('is_stable'))\
                       .orderBy(col('market_cap').desc())

df_gold_13.display()

df_gold_13.write.format("delta").mode('overwrite').saveAsTable('crypto.crypto_schema.gold_stablecoin_analysis')
```
